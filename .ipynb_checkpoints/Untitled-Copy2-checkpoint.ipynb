```python
import sys
import os
home = os.environ['HOME']
sys.path.append(home + '/projects/turbubox/tools/lib')
sys.path.append(home + '/projects/turbubox/tools')

import numpy as np
import pandas as pd
import glob
import cubicle
import turbubox.ulz as ulz

import matplotlib
matplotlib.rcParams.update({'font.size': 9})

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#fp = '/mnt/sshfs/supermuc2/gss/scratch/pr62su/ri97lab8/StirTurb_CGM_dense5_512/StirTurb_hdf5_plt_cnt_0013'
fp = '/home/jmark/Downloads/StirTurb_hdf5_plt_cnt_0013'
fh = cubicle.File(fp)
```

```python
dens = fh.get_data('dens', rlevel=None)
```

```python
temp = fh.get_data('temp', rlevel=None)
```

```python
dpi = 100
fig = plt.figure(1,figsize=(2*640/dpi, 480/dpi), dpi=dpi)
    
plt.title('density')
plt.imshow(
    np.log10(np.sum(dens,axis=2)),
    #extent = extent,
    #vmin = 0.0,
    #vmax = 1.2,
    origin='lower',
    interpolation = None,
    cmap = plt.get_cmap('cubehelix'),
)
plt.colorbar()
```

```python
dpi = 100
fig = plt.figure(1,figsize=(2*640/dpi, 480/dpi), dpi=dpi)
    
plt.title('temperature')
plt.imshow(
    #np.log10(np.sum(temp,axis=2)),
    np.log10(temp[:,:,temp.shape[2]//4]),
    #extent = extent,
    #vmin = 0.0,
    #vmax = 1.2,
    origin='lower',
    interpolation = None,
    cmap = plt.get_cmap('cubehelix'),
)
plt.colorbar()
```
