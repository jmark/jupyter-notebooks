```python
def harmonics(n,m,theta,phi):
    return scp.sph_harm(m, n, phi, theta)

def mk_multipole_moments(charges, radii, thetas, phis):
    return lambda n,m: np.sum(charges * radii**n * harmonics(n,-m,thetas,phis))
```

```python
if 1:
    ncells = 31
    npoly = 6
    Nvisu = npoly+1
    
    #nodes = gausslobatto.mk_nodes(npoly,ntype='gauss')
    nodes = ulz.mk_body_centered_linspace(-1,1,npoly+1)

    xs = ulz.mk_body_centered_linspace(0,1,ncells)
    Xs,Ys = np.meshgrid(xs,xs)

    MM = np.zeros_like(Xs)

    #MM += 1

    MM[len(MM)//2,len(MM)//2] = 0.8
    MM[2,2] = 0.5
    MM[-5,4] = 0.7

    dens = ulz.zoom_array(MM,npoly+1)
    
    # tiled radii and angles
    fs = np.linspace(0,1,ncells,endpoint=False)
    xs = np.array([nodes / ncells / 2 for y in fs])
    xs = np.ravel(xs)

    Xs,Ys  = np.meshgrid(xs,xs)
    radii  = np.sqrt(Xs**2+Ys**2)
    phis   = np.arctan2(Ys,Xs)
    thetas = np.pi/2

    charges = np.array([np.vsplit(col,ncells) for col in np.hsplit(dens,ncells)])
    radii   = np.array([np.vsplit(col,ncells) for col in np.hsplit(radii,ncells)])
    phis    = np.array([np.vsplit(col,ncells) for col in np.hsplit(phis,ncells)])
    
    MM     = np.array([[mk_multipole_moments(c,r,thetas,p)(0,0) for c,r,p in zip(crow,rrow,prow)] for crow,rrow,prow in zip(charges,radii,phis)])
    MP     = np.array([[mk_multipole_moments(c,r,thetas,p)(1,1) for c,r,p in zip(crow,rrow,prow)] for crow,rrow,prow in zip(charges,radii,phis)])
    #MPM    = np.array([[mk_multipole_moments(c,r,thetas,p)(1,-1) for c,r,p in zip(crow,rrow,prow)] for crow,rrow,prow in zip(charges,radii,phis)])
    
if 0:
    ncells = dens.shape[0]//Nvisu
    npoly  = 7
    nodes = ulz.mk_body_centered_linspace(-1,1,Nvisu)

    # tiled radii and angles
    fs = np.linspace(0,1,ncells,endpoint=False)
    xs = np.array([nodes / ncells / 2 for y in fs])
    xs = np.ravel(xs)

    Xs,Ys  = np.meshgrid(xs,xs)
    radii  = np.sqrt(Xs**2+Ys**2)
    phis   = np.arctan2(Ys,Xs)
    thetas = np.pi/2

    charges = np.array([np.vsplit(col,ncells) for col in np.hsplit(dens,ncells)])
    radii   = np.array([np.vsplit(col,ncells) for col in np.hsplit(radii,ncells)])
    phis    = np.array([np.vsplit(col,ncells) for col in np.hsplit(phis,ncells)])
    
    MM     = np.array([[mk_multipole_moments(c,r,thetas,p)(0,0) for c,r,p in zip(crow,rrow,prow)] for crow,rrow,prow in zip(charges,radii,phis)])
    MP     = np.array([[mk_multipole_moments(c,r,thetas,p)(1,1) for c,r,p in zip(crow,rrow,prow)] for crow,rrow,prow in zip(charges,radii,phis)])
    MPM    = np.array([[mk_multipole_moments(c,r,thetas,p)(1,-1) for c,r,p in zip(crow,rrow,prow)] for crow,rrow,prow in zip(charges,radii,phis)])

#rmin = 1/ncells/(npoly+1)*1
rmin = 0.05

def mk_resp_func(j,n,k,m):
    xs = ulz.mk_body_centered_linspace(0,1,ncells)
    Xs,Ys = np.meshgrid(xs,xs)

    def auxfun(n,m):
        return (-1)**n/np.sqrt(scipy.special.factorial(n-m)*scipy.special.factorial(n+m))
    
    def resp(x,y):
        radii  = np.sqrt((Xs-x)**2 + (Ys-y)**2)
        thetas = np.arccos(0.0)
        phis   = np.arctan2(Ys-y,1e-16+Xs-x)
        
        a = (1j)**(abs(k-m) - abs(k) - abs(m)) * auxfun(n,m) * auxfun(j,k) * harmonics(j+n,m-k,thetas, phis)
        b = (-1)**j * auxfun(j+n,m-k) * radii**(j+n+1)
        
        retval = a/(1e-16+b)
        retval[radii <= rmin] = 0.0
        return retval
    
    return resp

fft  = lambda x: np.fft.fftn(x)
ifft = lambda x: np.fft.ifftn(x)

def mk_far_PP(rMM,rMP,rMPM=None):
    TT = dict()
    LL = dict()
    MM = dict()
    PP = dict()

    TT[0,0, 0, 0] = mk_resp_func(0,0, 0, 0)

    TT[0,1, 0,-1] = mk_resp_func(0,1, 0,-1)
    TT[0,1, 0, 1] = mk_resp_func(0,1, 0, 1)
    
    TT[1,0,-1, 0] = mk_resp_func(1,0,-1, 0)
    TT[1,0, 1, 0] = mk_resp_func(1,0, 1, 0)
    
    TT[1,1,-1, 1] = mk_resp_func(1,1,-1, 1)
    TT[1,1, 1,-1] = mk_resp_func(1,1, 1,-1)

    TT[1,1,-1,-1] = mk_resp_func(1,1,-1,-1)
    TT[1,1, 1, 1] = mk_resp_func(1,1, 1, 1)
    
    MM[0,0]  = fft(rMM)
    MM[1,1]  = fft(rMP)
    #MM[1,-1] = fft(rMPM)
    MM[1,-1] = fft(-np.conjugate(rMP))

    xs = ulz.mk_body_centered_linspace(0,1,ncells)
    
    LL[0, 0] = np.zeros((len(xs),len(xs)),dtype=np.complex128)
    LL[1,-1] = np.zeros((len(xs),len(xs)),dtype=np.complex128)
    LL[1, 1] = np.zeros((len(xs),len(xs)),dtype=np.complex128)
    
    LL[0, 0] += np.array([[ifft(MM[0, 0] * fft(TT[0,0,0, 0](x,y)))[-1,-1] for x in xs] for y in xs])
    
    #LL[0, 0] += np.array([[ifft(MM[1,-1] * fft(TT[0,1,0,-1](x,y)))[-1,-1] for x in xs] for y in xs])
    LL[0, 0] += np.array([[ifft(MM[1, 1] * fft(TT[0,1,0, 1](x,y)))[-1,-1] for x in xs] for y in xs])

    #LL[1,-1] += np.array([[ifft(MM[0, 0] * fft(TT[1,0,-1, 0](x,y)))[-1,-1] for x in xs] for y in xs])
    LL[1, 1] += np.array([[ifft(MM[0, 0] * fft(TT[1,0, 1, 0](x,y)))[-1,-1] for x in xs] for y in xs])
    
    #LL[1,-1] += np.array([[ifft(MM[1,-1] * fft(TT[1,1,-1,-1](x,y)))[-1,-1] for x in xs] for y in xs])
    LL[1, 1] += np.array([[ifft(MM[1, 1] * fft(TT[1,1, 1, 1](x,y)))[-1,-1] for x in xs] for y in xs])
    
    #LL[1,-1] += np.array([[ifft(MM[1, 1] * fft(TT[1,1,-1, 1](x,y)))[-1,-1] for x in xs] for y in xs])
    LL[1, 1] += np.array([[ifft(MM[1,-1] * fft(TT[1,1, 1,-1](x,y)))[-1,-1] for x in xs] for y in xs])
   

    # tiled radii and angles
    nodes = ulz.mk_body_centered_linspace(-1,1,npoly+1)
    fs = np.linspace(0,1,ncells,endpoint=False)
    xs = np.array([nodes / ncells / 2 for y in fs])
    xs = np.ravel(xs)

    Xs,Ys  = np.meshgrid(xs,xs)
    radii  = np.sqrt(Xs**2+Ys**2)
    phis   = np.arctan2(Ys,Xs)
    thetas = np.pi/2
    
    PP[0, 0] = harmonics(0, 0,thetas,phis) * ulz.zoom_array(LL[0, 0], factor=npoly+1)
    PP[1,-1] = harmonics(1,-1,thetas,phis) * ulz.zoom_array(LL[1,-1], factor=npoly+1)
    PP[1, 1] = harmonics(1, 1,thetas,phis) * ulz.zoom_array(LL[1, 1], factor=npoly+1)
    
    #PP = PP[0,0] + 1/np.pi * radii * (PP[1,-1] + PP[1,1])
    
    PP = PP[0,0] + 2/np.pi * radii * PP[1,1]
    
    return PP
    #return np.flipud(np.fliplr(PP))
    
def mk_near_PP(MM):
    return -MM * 22

fig = plt.figure(1,figsize=(geo[0]/dpi, geo[1]/dpi), dpi=dpi)
plt.imshow(MM.real)
plt.colorbar()
```
